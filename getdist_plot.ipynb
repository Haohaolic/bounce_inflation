```python
from getdist import plots, loadMCSamples
from getdist.types import ResultTable

root_dir = '/home/li/packages/montepython_public/chains/'
dir1 = 'BI_Planck18_Al/Planck18+$A_L$'
dir2 = 'BI_Planck18_spt3g_Al/BI_Planck18_spt3g_Al'
dir3 = 'BI_Planck18/Planck18+$\Lambda$CDM'
dir4 = 'BI_Planck18_spt3g/Planck18+SPT3G+$\Lambda$CDM'
dir5 = 'BI_Planck18_Alt/BI_Planck18_Alt'
dir6 = 'BI_Planck18_spt3g_Alt/BI_Planck18_spt3g_Alt'

#samples = loadMCSamples(file_root = dir1+'2024-05-05_4000000_', ini = dir1+'Hl.ini')
samples1 = loadMCSamples(file_root = root_dir + dir1, settings={'ignore_rows': '0.3'})
samples2 = loadMCSamples(file_root = root_dir + dir2, settings={'ignore_rows': '0.3'})
samples3 = loadMCSamples(file_root = root_dir + dir3, settings={'ignore_rows': '0.3'})
samples4 = loadMCSamples(file_root = root_dir + dir4, settings={'ignore_rows': '0.3'})
samples5 = loadMCSamples(file_root = root_dir + dir5, settings={'ignore_rows': '0.3'})
samples6 = loadMCSamples(file_root = root_dir + dir6, settings={'ignore_rows': '0.7'})
```
